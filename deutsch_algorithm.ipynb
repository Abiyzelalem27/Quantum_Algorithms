## Oracle / Black box function (Uf)

Uf will be our black box, or oracle function. To do this, we will need to extend our previous diagram to include the extra components necessary to create our oracle:

1. First, we will convert our input and output registers into Dirac ket notation, |x⟩.
2. Next, we will create two input registers, |x⟩ and |y⟩, where the input registers will feed into our black box, or oracle function, Uf, and the |y⟩ register is used as an ancillary qubit.  
   Ancilla qubits are extra bits used to store information that might be used later or to track information throughout the quantum circuit.
3. Finally, we’ll define our two output registers: one that is just the same as the input |x⟩, and the other that is the XOR of the input register y and the function f(x), as |x, y ⊕ f(x)⟩.

Therefore, we can now define the oracle function as follows:

Uf : |x⟩, |y⟩ → |x⟩, |y ⊕ f(x)⟩

```python
import numpy as np

# -----------------------------
# Basic quantum helpers
# -----------------------------

def ket0():
    return np.array([[1],[0]], dtype=complex)

def ket1():
    return np.array([[0],[1]], dtype=complex)

def kron(a, b):
    return np.kron(a, b)

# Gates
I = np.eye(2, dtype=complex)

H = (1/np.sqrt(2)) * np.array([
    [1,  1],
    [1, -1]
], dtype=complex)

X = np.array([
    [0, 1],
    [1, 0]
], dtype=complex)

# -----------------------------
# Oracle Uf : |x,y> -> |x, y ⊕ f(x)>
# We will create Uf for different f(x)
# -----------------------------

def oracle_Uf(f0, f1):
    """
    f0 = f(0)  (0 or 1)
    f1 = f(1)  (0 or 1)

    Basis order: |00>, |01>, |10>, |11>
    where first qubit = x, second qubit = y
    """
    Uf = np.zeros((4,4), dtype=complex)

    basis = [
        (0,0), (0,1), (1,0), (1,1)
    ]

    for i, (x, y) in enumerate(basis):
        fx = f0 if x == 0 else f1
        y_new = y ^ fx

        # Find output basis index
        j = basis.index((x, y_new))
        Uf[j, i] = 1

    return Uf

# -----------------------------
# Deutsch Algorithm Simulation
# -----------------------------

def deutsch_algorithm(f0, f1):
    # Step 1: initialize |x,y> = |0,1>
    psi0 = kron(ket0(), ket1())

    # Step 2: apply H on both qubits
    psi1 = kron(H, H) @ psi0

    # Step 3: apply oracle Uf
    Uf = oracle_Uf(f0, f1)
    psi2 = Uf @ psi1

    # Step 4: apply H on first qubit only (input qubit)
    psi3 = kron(H, I) @ psi2

    # Measurement probabilities of x (first qubit)
    # Basis order: |00>, |01>, |10>, |11>
    probs = np.abs(psi3.flatten())**2

    p_x0 = probs[0] + probs[1]
    p_x1 = probs[2] + probs[3]

    return psi3, p_x0, p_x1

# -----------------------------
# Test Cases
# -----------------------------

# Constant: f(0)=0, f(1)=0
state, p0, p1 = deutsch_algorithm(0, 0)
print("Constant f(0)=0 f(1)=0")
print("P(x=0) =", p0, " P(x=1) =", p1)

# Constant: f(0)=1, f(1)=1
state, p0, p1 = deutsch_algorithm(1, 1)
print("\nConstant f(0)=1 f(1)=1")
print("P(x=0) =", p0, " P(x=1) =", p1)

# Balanced: f(0)=0, f(1)=1
state, p0, p1 = deutsch_algorithm(0, 1)
print("\nBalanced f(0)=0 f(1)=1")
print("P(x=0) =", p0, " P(x=1) =", p1)

# Balanced: f(0)=1, f(1)=0
state, p0, p1 = deutsch_algorithm(1, 0)
print("\nBalanced f(0)=1 f(1)=0")
print("P(x=0) =", p0, " P(x=1) =", p1)
```
