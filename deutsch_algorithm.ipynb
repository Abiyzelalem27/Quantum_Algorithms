Quantum Algorithm

- A Quantum Algorithm is step by step procedure designed for fun quantum computer, under uses the principle of QTM such as
{ Superposition, Interference, Entanglement } to some problems.

All quantum Algorithms work with the QMF Framework

1) All of them always initialized by qubit
2) Apply quantum gate/unitary operation to create superposition to manipulate the qubit
3) Oracle queries (Black box) – when is determinant
4) Following the above apply operation to break the superposition
5) Completed the algorithms and measures the qubit
   if measured qubit then completes Algo / our conclusion


Procedure/steps for apply Deutsch’s Algorithm

1- First put the input x and y in Dirac ket notation
2- 1st – apply Hadamard gate
3- 2nd – the input register in superposition state feed into black box or oracle function Uf and the ancilla register is used as auxiliary qubit
   - Ancilla qubit are extra bit used to store info that might be use later on to trick them throughout the quantum circuit
4- 4th – apply H – to break the superposition after oracle
5- Complete the Algorithm
6- Measured the qubit

Oracle / Black box function (Uf)

Uf will be our black box, or oracle function. To do this, we will need to extend our previous diagram to include the extra components necessary to create our oracle:

1. First, we will convert our input and output registers into Dirac ket notation, |x⟩.
2. Next, we will create two input registers, |x⟩ and |y⟩, where the input registers will feed into our black box, or oracle function, Uf, and the |y⟩ register is used as an ancillary qubit.  
   Ancilla qubits are extra bits used to store information that might be used later or to track information throughout the quantum circuit.
3. Finally, we’ll define our two output registers: one that is just the same as the input |x⟩, and the other that is the XOR of the input register y and the function f(x), as |x, y ⊕ f(x)⟩.

Therefore, we can now define the oracle function as follows:

Uf : |x⟩, |y⟩ → |x⟩, |y ⊕ f(x)⟩
